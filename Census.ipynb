# 4 - Divisão entre Previsores e Classe

```python
dados.columns
```

```python
#Previsores
x_census = dados[['age', 'workclass', 'fnlwgt', 'education', 'education-num',
       'marital-status', 'occupation', 'relationship', 'race', 'sex',
       'capital-gain', 'capital-loss', 'hours-per-week', 'native-country']].values
```

```python
x_census
```

```python
#Classe
y_census = dados['income'].values
```

```python
y_census
```

# 5 - LabelEncoder para os atributos Categóricos

```python
from sklearn.preprocessing import LabelEncoder
```

```python
x_census[0]
```

```python
label_encoder_workclass = LabelEncoder()
label_encoder_education = LabelEncoder()
label_encoder_marital = LabelEncoder()
label_encoder_occupation = LabelEncoder()
label_encoder_relationship = LabelEncoder()
label_encoder_race = LabelEncoder()
label_encoder_sex = LabelEncoder()
label_encoder_country = LabelEncoder()
```

```python
x_census[:,1] = label_encoder_workclass.fit_transform(x_census[:,1])
x_census[:,3] = label_encoder_education.fit_transform(x_census[:,3])
x_census[:,5] = label_encoder_marital.fit_transform(x_census[:,5])
x_census[:,6] = label_encoder_occupation.fit_transform(x_census[:,6])
x_census[:,7] = label_encoder_relationship.fit_transform(x_census[:,7])
x_census[:,8] = label_encoder_race.fit_transform(x_census[:,8])
x_census[:,9] = label_encoder_sex.fit_transform(x_census[:,9])
x_census[:,13] = label_encoder_country.fit_transform(x_census[:,13])
```

```python
x_census
```

# 6 - OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
onehotencoder_census = ColumnTransformer(transformers=[('OneHot', 
                                                        OneHotEncoder(), 
                                                        [1,3,5,6,7,8,9,13])], 
                                                        remainder='passthrough')
```

```python
x_census_onehotencoder = onehotencoder_census.fit_transform(x_census).toarray()
```

```python
x_census_onehotencoder.shape
```

# 7 - Escalonamento dos Valores

```python
from sklearn.preprocessing import StandardScaler
scaler_census = StandardScaler()
x_census_onehotencoder_scale = scaler_census.fit_transform(x_census_onehotencoder)
```

```python
x_census_onehotencoder_scale
```

```python

#Lembrete das variáveis definidas:

# x_census - base de dado de previsores somento com o tratamento de label encoder;
# x_census_onehotencoder - base de dados de previsores com o tratamento de labelencoder e onehotencoder;
# x_census_onehotencoder_scale - base de dados de previsores com tratmento de labelencoder,onehotencoder e escalonamento.


```

# 8 - Divisão entre Base de Dados de Teste e Base de Dados de Treinamento

```python
from sklearn.model_selection import train_test_split
```

```python
#x_census e y_census
x_census_treinamento,x_census_teste, y_census_treinamento, y_census_teste = train_test_split(x_census, y_census, test_size = 0.15, random_state = 0)
```

```python
#x_census_onehotencoder
x_census_onehotencoder_treinamento,x_census_onehotencoder_teste, y_census_onehotencoder_treinamento,y_census_onehotencoder_teste = train_test_split(x_census_onehotencoder,y_census,test_size=0.15,random_state=0)
```

```python
#x_census_onehotencoder_scale
x_census_onehotencoder_scale_treinamento,x_census_onehotencoder_scale_teste, y_census_onehotencoder_scale_treinamento,y_census_onehotencoder_scale_teste = train_test_split(x_census_onehotencoder_scale,y_census,test_size=0.15,random_state=0) 
```

```python
#Salvando as variaveis
import pickle 
with open('census.pkl', mode = 'wb') as f:
  pickle.dump([x_census_treinamento, y_census_treinamento, x_census_teste, y_census_teste], f)
```

```python
with open('census_onehotencoder.pkl', mode = 'wb') as f:
  pickle.dump([x_census_onehotencoder_treinamento,  y_census_onehotencoder_treinamento, x_census_onehotencoder_teste, y_census_onehotencoder_teste], f)
```

```python
with open('census_onehotencoder_scale.pkl', mode = 'wb') as f:
  pickle.dump([x_census_onehotencoder_scale_treinamento,  y_census_onehotencoder_scale_treinamento, x_census_onehotencoder_scale_teste, y_census_onehotencoder_scale_teste], f)
```

# 9 - Implemntanção do Algoritmo de Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
#Base de Dados só com tratamento de LabelEncoder

x_census_treinamento.shape, y_census_treinamento.shape
```

```python
random_forest_census = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state = 0)
random_forest_census.fit(x_census_treinamento, y_census_treinamento)
```

```python
previsoes_census = random_forest_census.predict(x_census_teste)
```

```python
accuracy_score(y_census_teste,previsoes_census)
```

```python
print(classification_report(y_census_teste, previsoes_census))
```

```python
#Base de Dados  co tratamento de LabelEncoder e OneHotEncoder
x_census_onehotencoder_treinamento.shape, y_census_onehotencoder_scale_treinamento.shape
```

```python
random_forest_census_onehotencoder = RandomForestClassifier(n_estimators=120, criterion='entropy', random_state = 0)
random_forest_census_onehotencoder.fit(x_census_onehotencoder_treinamento, y_census_onehotencoder_treinamento)
```

```python
previsoes_census_onehotencoder = random_forest_census_onehotencoder.predict(x_census_onehotencoder_teste)
```

```python
accuracy_score(y_census_onehotencoder_teste,previsoes_census_onehotencoder)
```

```python
print(classification_report(y_census_onehotencoder_teste, previsoes_census_onehotencoder))
```

```python
#Base de Dados  co tratamento de LabelEncoder,OneHotEncoder e Escalonmaneto dos Valores
x_census_onehotencoder_scale_treinamento.shape, y_census_onehotencoder_scale_treinamento.shape
```

```python
random_forest_census_onehotencoder_scale = RandomForestClassifier(n_estimators=120, criterion='entropy', random_state = 0)
random_forest_census_onehotencoder_scale.fit(x_census_onehotencoder_scale_treinamento,y_census_onehotencoder_scale_treinamento)
```

```python
previsoes_census_onehotencoder_scale = random_forest_census_onehotencoder_scale.predict(x_census_onehotencoder_scale_teste)
```

```python
accuracy_score(y_census_onehotencoder_scale_teste,previsoes_census_onehotencoder_scale)
```

```python
print(classification_report(y_census_onehotencoder_scale_teste, previsoes_census_onehotencoder_scale))
```
